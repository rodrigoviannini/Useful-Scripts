```python
import os

def files_path05(*args):
    ''' Função que percorre todas as pastas até os arquivos de imagens '''
    for item in args:
        for p, _, files in os.walk(os.path.abspath(item)):
            for file in files:
                print(os.path.join(p, file))
                # print(file)
```

```python
def chr_remove(old, to_remove):
    new_string = old
    for x in to_remove:
        new_string = new_string.replace(x, '')
    return new_string
```
